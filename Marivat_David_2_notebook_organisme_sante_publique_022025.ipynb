## Analyse du boxplot de la variable omega-3-fat_100g

### 1. Mesure de tendance centrale  
La médiane de cette distribution est de **1.80g/100g**, indiquant le point central des données.  
La moyenne est de **3.18g/100g**. Si la moyenne est **supérieure** à la médiane, la distribution est potentiellement **asymétrique à droite**. À l’inverse, si elle est **inférieure**, l’asymétrie est **à gauche**.

### 2. Mesure de dispersion  
L’écart-type est de **5.61g/100g**, ce qui représente la dispersion des valeurs autour de la moyenne.

Les moustaches du boxplot sont définies par la **méthode interquartile** :  
\[
[Q1 - 1.5 \times IQR, Q3 + 1.5 \times IQR]
\]
où \( IQR = Q3 - Q1 \).  
Les valeurs situées au-delà de cette plage sont considérées comme **outliers (valeurs aberrantes)**.

### ⚠️ Impact sur la Santé d’un Apport Insuffisant ou Excessif en Oméga-3  

#### ✔️ Manque d’oméga-3 : Risques pour la Santé  
Un apport insuffisant en oméga-3 peut entraîner :  
- **Un risque accru de maladies cardiovasculaires**, car les oméga-3 participent à la régulation du cholestérol et de la pression artérielle.  
- **Des troubles cognitifs et une baisse de la concentration**, car ces acides gras sont essentiels au bon fonctionnement du cerveau.  
- **Une inflammation accrue**, augmentant les douleurs articulaires et certains troubles inflammatoires chroniques.  

#### ✔️ Excès d’oméga-3 : Effets Potentiellement Négatifs  
Un apport excessif, bien que rare, peut :  
- **Fluidifier excessivement le sang**, augmentant le risque d’hémorragie en cas de prise d’anticoagulants.  
- **Causer des troubles digestifs** (diarrhée, reflux).
