```python
# Afficher un scatter plot entre 'omega-3-fat_100g' et 'omega-6-fat_100g'
test_spearman(nutritional_products, 'omega-3-fat_100g', 'omega-6-fat_100g')
```

### **📝 Modèle d’Analyse d’une Heatmap de Corrélation**  

📌 **Analyse de la relation entre `{omega-3-fat_100g}` et `{omega-6-fat_100g}`**  

🔍 **Valeurs de corrélation (Spearman)**  
- **Corrélation `{omega-3-fat_100g}` avec `{omega-3-fat_100g}`** → **1 (Rouge)** ✅  
- **Corrélation `{omega-3-fat_100g}` avec `{omega-6-fat_100g}`** → **{0.65}**  
- **Corrélation `{omega-6-fat_100g}` avec `{omega-3-fat_100g}`** → **{0.65}**  
- **Corrélation `{omega-6-fat_100g}` avec `{omega-6-fat_100g}`** → **1 (Rouge)** ✅  

📈 **Type et intensité de la corrélation**   
  - **Modérée** (0.3 à 0.7) → Relation significative.     

🧐 **Interprétation possible**  
- **Une corrélation modérée entre `{omega-3-fat_100g}` et `{omega-6-fat_100g}`** suggère qu'une variable pourrait influencer l'autre ou qu'elles ont une cause commune.    

✅ **Conclusion**  
En observant la heatmap, nous pouvons conclure que **{omega-3-fat_100g} et {omega-6-fat_100g}** ont une corrélation **{modérée}** et **{positive}**, avec une valeur de **{0.65}**.  
Cela signifie que **{ lorsque la teneur en oméga-3 augmente, la teneur en oméga-6 a tendance à augmenter aussi dans les produits analysés}**.
